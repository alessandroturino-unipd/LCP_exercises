1\. **Radioactive decay chain**

${\rm Tl}^{208}$ decays to ${\rm Pb}^{208}$ with a half-lieve of 3.052 minutes. Suppose to start with a sample of 1000 Thallium atoms and 0 of Lead atoms.

* Take steps in time of 1 second and at each time-step decide whether each Tl atom has decayed or not, accordingly to the probability $p(t)=1-2^{-t/\tau}$. Subtract the total number of Tl atoms that decayed at each step from the Tl sample and add them to the Lead one. Plot the evolution of the two sets as a function of time  
* Repeat the exercise by means of the inverse transform method: draw 1000 random numbers from the non-uniform probability distribution $p(t)=2^{-t/\tau}\frac{\ln 2}{\tau}$ to represent the times of decay of the 1000 Tl atoms. Make a plot showing the number of atoms that have not decayed as a function of time

```python
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt

# the simulation is in secs

np.random.seed(100)
N_atom=1000
time_step=np.arange(0,1000) # time step 1 sec for 1000 sec

Tl_atom=np.zeros(len(time_step))
Tl_atom[0]=N_atom

Pb_atom=np.zeros(len(time_step))
Pb_atom[0]=0

t_half=3.052 * 60 # half-life costant in sec
tau=t_half/np.log(2) 

dt = time_step[1] - time_step[0]
p=1-np.exp(-dt/tau)
for i in range(1,len(time_step)):
    Tl_atom[i]=Tl_atom[i-1]
    Pb_atom[i]=Pb_atom[i-1]
    for j in range(int(Tl_atom[i-1])):
        r=npr.random()
        if r < p:
            Tl_atom[i]-=1
            Pb_atom[i]+=1

plt.plot(time_step,Tl_atom,color='red',label='Thallium population')
plt.plot(time_step,Pb_atom,color='blue',label='Lead population')
plt.legend()
plt.title('Thallium decay')
```

```python
## point 2
x=-tau*np.log(1-npr.random(N_atom))
x=np.sort(x)

Tl_atom1=np.zeros(len(time_step))
Tl_atom1[0]=N_atom

Pb_atom1=np.zeros(len(time_step))
Pb_atom1[0]=0
for i in range(1,len(time_step)):
     time=time_step[i-1]
     decay=0
     for j in range(N_atom):
             if x[j]<time:
                 decay+=1
             
     
     Tl_atom1[i]=N_atom-decay        
    
plt.plot(time_step,Tl_atom1,color='red',label='Survivor population')
plt.legend()
```

2\. **Rutherford Scattering**

The scattering angle $\theta$ of $\alpha$ particles hitting a positively charged nucleus of a Gold atom ($Z=79$) follows the rule:

$$
\tan{\frac{1}{2} \theta} = \frac{Z e^2} {2\pi \epsilon_0 E b}
$$

where $E=7.7$ MeV and $b$ beam is the impact parameter. The beam is represented by a 2D gaussian distribution with $\sigma=a_0/100$ for both coordinates ($a_0$ being the Bohr radius). Assume 1 million $\alpha$ particles are shot on the gold atom.

Computing the fraction of particles that "bounce back",i.e. those particle whose scattering angle is greater than $\pi/2$ (which set a condition on the impact parameter $b$)

```python
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt

#np.random.seed(1)
alpha=1/137
a_0=1./alpha # natural units
Z=79
e_0=1 # natural units
e=1 # natural units
hc=197 # Mev * fm
E=7.7/hc # natural units
Num=1000000 # number of particles
mns=[0,0]
COV=[[a_0/100.0,0],[0,a_0/100.]]
data=npr.multivariate_normal(mns,COV,int(Num))
#help(npr.multivariate_normal)
l_angle=np.pi/2
theta=np.arctan2(data[:,1],data[:,0])
theta=np.sort(theta)
Num_sim=1 # number of simulation
part_frac=np.zeros(Num_sim)
for j in range(Num_sim):
    for i in range(Num):
        if abs(theta[i])>l_angle:
            part_frac[j]+=1


b=Z*e/(np.tan(theta/2)*2*np.pi*e_0*E)/1e11 #nanometer
part_frac/=Num  
#plt.plot(b,part_frac)
#plt.hist(part_frac,200)
#help(plt.hist)
```

```python
import numpy as np
import numpy.random as npr

Num = 1_000_000
Z = 79
E = 7.7          # MeV
const = 1.44     # e^2/(40) in MeVfm

# Bohr radius in fm
a0_fm = 5.29e4
sigma = a0_fm / 100

# Gaussian beam
data = npr.multivariate_normal([0,0],
                               [[sigma**2,0],[0,sigma**2]],
                               Num)
x, y = data[:,0], data[:,1]
b = np.sqrt(x*x + y*y)

# critical impact parameter for  = /2
b_crit = (Z * const) / (2 * E)

fraction = np.mean(b < b_crit)

print("b_crit =", b_crit, "fm")
print("Beam sigma =", sigma, "fm")
print("Backscatter fraction =", fraction)
```

4\. **Monte Carlo integration in high dimension**

* Start of by computing the area of a circle of unit radius, by integrating the function 

$$
f(x,y)=
\left\{
\begin{array}{ll}
      1 & x^2+y^2\le 1 \\
      0 & {\rm elsewhere}
\end{array} 
\right.
$$

* Generalize the result for a 10D sphere



```python
import numpy as np
import numpy.random as npr

npr.seed(1)
def norm(x):
    f = lambda x : np.sqrt(sum(i*i for i in x))
    return f(x)

N=10000 #number of points
# first in 2d
x=npr.random(size=N)
y=npr.random(size=N)
count=0
for i in range(N):
    if norm([x[i],y[i]])<=1:
        count+=1


A=2**2 # area of the 2 power number of dimensions
I=count/N *A
print('The estimated area in 2D is :', I)

# now for 10D
D=10
A=2**D
xx=np.zeros(D)
count=0
for i in range(N):
    for j in range(D):
        xx[j]=npr.random()  
    
    if norm(xx)<=1:
        count+=1
        
I=count/N *A
print('The estimated area in 10D is :', I)
```

5\. **Monte Carlo integration with importance sampling** 

Calculate the value of the integral:

$$
I=\int_0^1 \frac{x^{-1/2}}{e^x+1} dx
$$

using the importance sampling method with $w(x)=1/\sqrt{x}$. You should get a result about 0.84

```python
import numpy as np
import numpy.random as npr

def f(x):
    return 1./(np.exp(x)+1.)

A=2 # result of the integration of 1/sqrt(x)
N=10000
x=npr.random(size=N)
y=x*x # this comes from inverting the pdf from a U[0-1] to 1/(2*sqrt(x)) in [0,1]
g=f(y)
I=2./N *g.sum()
print('The estimated integral is :',I)
```
