5\. Use HOFs (zip in particular) to compute the weight of a circle, a disk and a sphere, assuming different radii and different densities:

```python
densities = {"Al":[0.5,1,2],"Fe":[3,4,5],"Pb": [15,20,30]}
radii = [1,2,3]
```

where the entries of the dictionary's values are the linear, superficial and volumetric densities of the materials respectively.

In particular define a list of three lambda functions using a comprehension that computes the circumference, the area and the volume for a given radius.


````python
from math import pi
densities = {"Al":[0.5,1,2],"Fe":[3,4,5],"Pb": [15,20,30]}
radii = [1,2,3]

circle = lambda x,y : 2*pi*x*y
disk = lambda x,y : 2*pi*x*x*y
sphere = lambda x,y : 4./3. *pi *x**3*y
for mat,den in densities.items():
    for d,r in zip(den,radii):
        print('Mass of:',mat,'circle, with density d=',d,' and radius r=',r,'is : ',circle(r,d))
        print('Mass of:',mat,'disk, with density d=',d,' and radius r=',r,'is : ',disk(r,d))
        print('Mass of:',mat,'sphere, with density d=',d,' and radius r=',r,'is : ',sphere(r,d))
````

````python
import scipy as sp
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

npr.seed(10)
N = 100
mu = 0
sigma = 2
x = npr.normal(loc=mu, scale=sigma, size=N)

fig, ax = plt.subplots(figsize=(10,8))

# Histogram
counts, bin_edges, _ = ax.hist(x, range=[-4*sigma,4*sigma], bins=20)
ax.set_ylabel('Random data')
ax.set_xlabel('bins')
ax.yaxis.set_major_locator(MultipleLocator(1))

# Error bars
errs = np.sqrt(counts)
bin_center = (bin_edges[1:] + bin_edges[:-1]) / 2
ax.errorbar(x=bin_center, y=counts, yerr=errs, color='red', fmt='none')

# Gaussian generator
def gaussian(mu, sigma):
    return sp.stats.norm(loc=mu, scale=sigma)

# Scott bandwidth
h = 1.06 * x.std() * x.size**(-1./5.)

# Build KDE
gauss = gaussian(x, h)
xgrid = np.linspace(-4*sigma, 4*sigma, N)

pdf = gauss.pdf(xgrid[:, np.newaxis])

# Plot individual Gaussians (optional, as in your original)
ax.plot(xgrid, pdf, zorder=1, alpha=0.1)

# Sum all Gaussians
kde_sum = np.sum(pdf, axis=1)

# --- Normalization step requested ---
# Integral of histogram:
hist_integral = np.sum(counts * np.diff(bin_edges))

# Integral of KDE curve:
kde_integral = sp.integrate.trapezoid(kde_sum, xgrid)

# Scale KDE to match histogram integral
kde_sum *= hist_integral / kde_integral
# ------------------------------------

# Plot normalized KDE
ax.plot(xgrid, kde_sum, color='red')

plt.show()
````
